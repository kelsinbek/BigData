## 2-Построить график произвольной квадратичной функции

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)  
y1 = x**2 

plt.figure(figsize=(8, 6)) 
plt.plot(x, y1, label='y = x^2')  
plt.xlabel('X') 
plt.ylabel('Y')  
plt.title('График квадратичной функции')  
plt.legend()  
plt.grid(True)  
plt.show() 
```

## 3 Построить график произвольной гармонической функции

```python
x = np.linspace(0, 10, 100)  
y2 = np.sin(x)  

plt.figure(figsize=(8, 6))  
plt.plot(x, y2, label='y = sin(x)') 
plt.xlabel('X') 
plt.ylabel('Y')  

plt.title('График гармонической функции')  
plt.legend()  
plt.grid(True)  
plt.show() 
```

## 4 Вывести графики из пунктов 2 и 3 в разные поля при помощи функции subplot()

```python
plt.figure(figsize=(12, 6))  

plt.subplot(1, 2, 1) 
plt.plot(x, y2, label='y = sin(x)')  
plt.xlabel('X')  
plt.ylabel('Y')  
plt.title('График гармонической функции')  
plt.legend()  
plt.grid(True)  

plt.subplot(1, 2, 2)  
plt.plot(x, y1, label='y = x^2')  
plt.xlabel('X')  
plt.ylabel('Y')  
plt.title('График квадратичной функции')  
plt.legend()  
plt.grid(True) 

plt.tight_layout()  
plt.show() 
```

## 5  Построить графики из пунктов 2 и 3 в одной координатной плоскости с легендой

```python
plt.figure(figsize=(8, 6)) 
plt.plot(x, y2, label='y = sin(x)')  
plt.plot(x, y1, label='y = x^2')  
plt.xlabel('X')  
plt.ylabel('Y')  
plt.title('Графики функций')  
plt.legend()  
plt.grid(True)  
plt.show()  
```

## 6 Построить график функции, заданной параметрически

```python
t = np.linspace(0, 4 * np.pi, 100)  
xt = np.cos(3 * t) 
yt = np.sin(2 * t)  

plt.figure(figsize=(8, 6)) 
plt.plot(xt, yt, label='x(t)=cos(3t), y(t)=sin(2t)') 
plt.xlabel('X(t)') 
plt.ylabel('Y(t)') 
plt.title('График параметрической функции') 
plt.legend()  
plt.grid(True)  
plt.show()  
```

## 7  Построить столбиковую диаграмму

```python

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
pr_min = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

plt.figure(figsize=(10, 6))  
plt.bar(years, pr_min, color='skyblue', alpha=0.7)  
plt.xlabel('Год')  
plt.ylabel('Прожиточный минимум') 
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')  
plt.xticks(years)  
plt.grid(True)  
plt.show()  
```

## 8  Построить круговой график

```python
plt.figure(figsize=(9, 9))  
labels = [str(year) for year in years]
plt.pie(pr_min, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Доля прожиточного минимума в Кыргызской Республике (2012-2022)')
plt.axis('equal')  
plt.show() 
```

## 9 Построить график произвольной каркасной поверхности

```python
X, Y = np.meshgrid(years, pr_min)
Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('Год')
ax.set_ylabel('Прожиточный минимум')
ax.set_zlabel('Значение функции')
ax.set_title('Каркасная поверхность')

plt.show()  
```
